```python
# Import libraries

import numpy as np

import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs

from sklearn.cluster import KMeans
 
# Step 1: Generate synthetic data

X, y_true = make_blobs(

    n_samples=500,      # number of points

    centers=4,          # true number of clusters

    cluster_std=0.60,   # cluster spread

    random_state=0

)
 
# Step 2: Plot the synthetic data

plt.figure(figsize=(7,5))

plt.scatter(X[:, 0], X[:, 1], s=50, color='gray')

plt.title("Synthetic Data Scatter Plot")

plt.xlabel("Feature 1")

plt.ylabel("Feature 2")

plt.show()
 
# Step 3: Use Elbow Method to find optimal k

inertia = []

K = range(1, 11)

for k in K:

    kmeans = KMeans(n_clusters=k, n_init=10, random_state=0)

    kmeans.fit(X)

    inertia.append(kmeans.inertia_)
 
# Step 4: Plot the Elbow Curve

plt.figure(figsize=(7,5))

plt.plot(K, inertia, 'bo-')

plt.title('Elbow Curve for Optimal k')

plt.xlabel('Number of clusters (k)')

plt.ylabel('Inertia')

plt.xticks(K)

plt.grid(True)

plt.show()
 
# Step 5: Apply KMeans with chosen k

k_opt = 4  # From elbow curve

kmeans = KMeans(n_clusters=k_opt, n_init=10, random_state=0)

y_kmeans = kmeans.fit_predict(X)
 
# Step 6: Visualize final clusters

plt.figure(figsize=(7,5))

plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_

plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, marker='X', label='Centroids')

plt.title(f'KMeans Clustering (k={k_opt})')

plt.xlabel("Feature 1")

plt.ylabel("Feature 2")

plt.legend()

plt.show()

 
```

```python
new_points = np.array([
    [0, 2],
    [1, 3],
    [2, 1],
    [4, 5],
    [6, 2],
    [3, 3],
    [5, 1]
])

# Predict their cluster labels
print(kmeans.predict(new_points))
```
