```python
import pandas as pd
```

```python
df = pd.read_csv('smart_meter_data.csv')
```

```python
df.head()
```

```python
df['location'].unique()
```

```python
locations = ['Chennai', 'Delhi', 'Bengaluru', 'Kolkata', 'Mumbai']
new_df = pd.concat(
    [df[df['location'] == loc].head(10000) for loc in locations],
    ignore_index=True
)
```

```python
new_df.sample(50)
```

```python
new_df.to_csv('transformed_smart_meter_data.csv', index=False)
```
