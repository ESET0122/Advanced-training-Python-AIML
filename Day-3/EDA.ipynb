```python
from sklearn.model_selection import train_test_split
Y = housing.Price
x = housing.drop(columns = ['Price'])
print(Y,x)
```

```python
X_train_full, X_valid_full, y_train, y_valid = train_test_split(x,Y,train_size = 0.8, random_state = 0)
```

```python
# Low cardinality categorical columns
low_cardinality_cols = [
    cname for cname in X_train_full.columns
    if X_train_full[cname].nunique() < 10 and X_train_full[cname].dtype == "object"
]

# Numerical columns
numerical_cols = [
    cname for cname in X_train_full.columns
    if X_train_full[cname].dtype in ['int64', 'float64']
]
```

```python
total_cols = low_cardinality_cols+numerical_cols
```

```python
## Actual training data for model
X_train = X_train_full[total_cols].copy()
X_valid = X_valid_full[total_cols].copy()
```

```python
cat = (X_train.dtypes == 'object')
list_cat = list(cat[cat].index)
list_cat
```

```python
from sklearn.preprocessing import OneHotEncoder

Oh_encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
Oh_cols_train = pd.DataFrame(Oh_encoder.fit_transform(X_train[list_cat]))
Oh_cols_valid = pd.DataFrame(Oh_encoder.fit_transform(X_valid[list_cat]))

Oh_cols_train.index = X_train.index
Oh_cols_valid.index = X_valid.index

num_x_train = X_train.drop(list_cat,axis=1)
num_x_valid = X_valid.drop(list_cat,axis=1)

oh_x_train = pd.concat([num_x_train,Oh_cols_train],axis=1)
oh_x_valid = pd.concat([num_x_valid,Oh_cols_valid],axis=1)

oh_x_train.columns = oh_x_train.columns.astype(str)
oh_x_valid.columns =  oh_x_valid.columns.astype(str)
```

## Handling Numeric Features
There are two ways of handling missing values

### 1. Removing Missing Values
### 2. Data Imputation

```python
car_data = pd.read_csv('D:\Advanced-training-Python-AIML\datasets\car.csv')
car_data.head(2)
```

```python
# Checking missing values
print(car_data.isnull().sum())
# Checking missing ratio
data_na = (car_data.isnull().sum()/len(car_data))*100
data_na
```

```python
# 1. groupby Method (fill the value with similar item of value in feature)
car_data['Engine Fuel Type'] = car_data.groupby('Model')['Engine Fuel Type'].transform(lambda x:x.fillna(x.mode().iloc[0] if not x.mode().empty else None))
car_data['Number of Doors'] = car_data.groupby('Model')['Number of Doors'].transform(lambda x:x.fillna(x.mean()))
car_data['Engine Cylinders'] = car_data.groupby('Model')['Engine Cylinders'].transform(lambda x:x.fillna(x.mean()))
car_data['Engine HP'] = car_data.groupby('Model')['Engine HP'].transform(lambda x:x.fillna(x.mean()))

car_data
```

```python
# 2. Rule-BasedÂ Imputation
null_data = car_data[car_data['Engine Cylinders'].isnull()]
null_data
```

```python
car_data.loc[(car_data['Engine Cylinders'].isna()) & (car_data['Engine Fuel Type']=='electric'), 'Engine Cylinders']=0
```
