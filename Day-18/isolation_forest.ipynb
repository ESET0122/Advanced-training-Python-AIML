```python
import pandas as pd
df = pd.read_csv('healthcare.csv')
print(df.head())
```

```python
from sklearn.ensemble import IsolationForest
```

```python
import matplotlib.pyplot as plt
plt.scatter(df.iloc[:,0], df.iloc[:,1])
```

```python
clf = IsolationForest(contamination=0.1)
clf.fit(df)
predictions = clf.predict(df)
```
